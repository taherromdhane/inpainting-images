```python
import numpy as np
np.set_printoptions(precision=3, suppress=True)
```

```python
patch_size = 3

img = np.random.rand(6, 6, 3)
print(img)
```

```python
def find_vector_set(img, patch_size):
    dims = img.shape
    vector_set = np.zeros((dims[0]*dims[1], 3 * patch_size**2))
    offset = patch_size//2
    
    for i in range(offset, dims[0] - offset - 1) :
        for j in range(offset, dims[1] - offset - 1) :
            
            block = img[i - offset : i + offset + 1, j - offset : j + offset + 1, :]
            feature = block.ravel()
            vector_set[dims[0] * i + j, :] = feature
            
    mean_vec = np.mean(vector_set, axis = 0)
    vector_set = vector_set - mean_vec 
            
    return vector_set
```

```python
vector_set = find_vector_set(img, patch_size)
```

```python
from sklearn.decomposition import PCA
 
pca = PCA()
pca.fit(vector_set)
EVS = pca.components_
importance = pca.explained_variance_ratio_
```
